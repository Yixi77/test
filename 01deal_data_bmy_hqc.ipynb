```python
import pandas as pd
o_hqc_d_bmy_result=pd.read_csv('o_hqc_d_bmy_result.csv',index_col=0)
o_hqc_d_bmy_result.head()
```

```python
# 获取所有列名的列表
column_names = o_hqc_d_bmy_result.columns.tolist()

# 找到要删除的列的索引
start_index = column_names.index('o_adcode')
end_index = column_names.index('d_y')

# 创建一个要删除的列名列表
columns_to_drop = column_names[start_index:end_index + 1]#取出的一个列表切片
# 删除这些列
o_hqc_d_bmy = o_hqc_d_bmy_result.drop(columns=columns_to_drop, inplace=False)
o_hqc_d_bmy.head()
```

```python
new_column_value = '华清池'
o_hqc_d_bmy.insert(1, '起点POI', new_column_value)

new_column_value = '兵马俑'
o_hqc_d_bmy.insert(2, '终点POI', new_column_value)
o_hqc_d_bmy.head()
```

```python
# 获取所有列名的列表
column_names = o_hqc_d_bmy.columns.tolist()

# 找到要删除的列的索引
start_index = column_names.index('geometry')
end_index = column_names.index('name_right')

# 创建一个要删除的列名列表
columns_to_drop = column_names[start_index:end_index + 1]#取出的一个列表切片
# 删除这些列
o_hqc_d_bmy = o_hqc_d_bmy.drop(columns=columns_to_drop, inplace=False)
o_hqc_d_bmy.head()
```

```python
o_hqc_d_bmy.columns=['日期','起点POI','终点POI','出发时间','到达时间','出行方式','出行时间','出行距离','方向','人数','置信度']
del o_hqc_d_bmy['方向']
o_hqc_d_bmy
```

```python
o_hqc_d_bmy=o_hqc_d_bmy.reset_index()
del o_hqc_d_bmy['index']
o_hqc_d_bmy
```

```python
o_hqc_d_bmy.to_csv('从华清池到兵马俑.csv')
```
